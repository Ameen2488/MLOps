## 1. The Broken YAML
The following YAML has 3 errors. Can you find them?
1. Wrong indentation for `steps`.
2. Missing `runs-on`.
3. `uses` keyword used for a cli command.

```yaml
name: Broken CI
on: push

jobs:
  test:
  # Error 2: Missing runs-on here
  steps:
    - uses: actions/checkout@v3
    - name: Run Tests
      # Error 3: 'uses' should be 'run'
      uses: pytest
```

````python
%%writefile fixed_ci.yaml
name: Fixed CI
on: push

jobs:
  test:
    runs-on: ubuntu-latest
    steps:
      - uses: actions/checkout@v3
      - name: Run Tests
        run: pytest
````
