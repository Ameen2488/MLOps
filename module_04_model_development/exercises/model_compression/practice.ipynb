# Exercise: Model Compression

**Objective**: Prune a PyTorch layer and observe weight sparsity.

---

## 1. Setup Layer

```python
import torch
import torch.nn as nn
import torch.nn.utils.prune as prune

layer = nn.Linear(5, 5)
print("Original Weights:\n", layer.weight)
```

## 2. Prune
Remove 30% of connections using `l1_unstructured`.

```python
# TODO: Prune
prune.l1_unstructured(layer, name="weight", amount=0.3)

print("Pruned Weights (Note the zeros):\n", layer.weight)
```

## 3. Permanent
Apply the mask permanently.

```python
prune.remove(layer, "weight")
print("Final Weights:\n", layer.weight)
```
