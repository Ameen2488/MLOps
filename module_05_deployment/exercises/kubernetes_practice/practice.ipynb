# Exercise: Kubernetes Practice

**Objective**: Create a Pod Manifest.

---

## 1. Create Manifest

```python
%%writefile pod.yaml
apiVersion: v1
kind: Pod
metadata:
  name: nginx-pod
spec:
  containers:
  - name: nginx
    image: nginx:alpine
```

## 2. Apply (Requires Minikube/K8s)
Run this in terminal.

```python
!kubectl apply -f pod.yaml
```
