## 1. Create Cluster Command
Assume we want a cluster named `demo` in `us-east-1` with `t3.small` nodes.

```python
# DRY RUN command
print("Command you would run:")
print("eksctl create cluster --name demo --region us-east-1 --node-type t3.small --nodes 2 --managed")
```

## 2. Update Kubeconfig
To access the cluster, you need to update your local kubeconfig.

```python
print("aws eks update-kubeconfig --name demo --region us-east-1")
```
