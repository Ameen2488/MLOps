# Exercise: Prefect Workflows

**Objective**: Define a basic Prefect Flow.

---

## 1. Define Tasks & Flow

```python
from prefect import task, flow

@task
def add(x, y):
    print(f"Adding {x} + {y}")
    return x + y

@task
def multiply(x, y):
    print(f"Multiplying {x} * {y}")
    return x * y

@flow
def math_flow(a, b, c):
    sum_res = add(a, b)
    final_res = multiply(sum_res, c)
    return final_res

if __name__ == "__main__":
    result = math_flow(1, 2, 3)
    print(f"Result: {result}")
```
