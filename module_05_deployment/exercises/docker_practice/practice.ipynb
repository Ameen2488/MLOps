# Exercise: Docker Practice

**Objective**: Create a Dockerfile and build an image.

---

## 1. Create Script

```python
%%writefile script.py
import sys
print(f"Hello from Python {sys.version}")
```

## 2. Create Dockerfile

```python
%%writefile Dockerfile
FROM python:3.9-slim
WORKDIR /app
COPY script.py .
CMD ["python", "script.py"]
```

## 3. Build & Run
