## 1. The Silent Killer of ML Models

In traditional software, code doesn't rot. `print("hello")` works today and in 10 years.
In ML, **models degrade** because the world changes.

> "Training data is a snapshot of the past. Inference data is the living present."

### The Two Main Types
1. **Data Drift (Covariate Shift)**: $P(X)$ changes.  
   *Example*: You trained on sunny images, but users start uploading night images. The input distribution shift.
2. **Concept Drift**: $P(Y|X)$ changes.  
   *Example*: A "spam" email in 2000 looks different from "spam" in 2024. The relationship between words and "spamness" flipped.

## 2. Data Drift (Covariate Shift)

The input features change distribution.
- **Seasonality**: Ice cream sales in winter vs summer.
- **New User Demographics**: App launches in a new country age group.
- **Broken Sensor**: A thermometer breaks and reports values +10 degrees higher.

## 3. Concept Drift

The *meaning* of the data changes.
- **Inflation**: A \$50,000 salary in 1990 was "High Income". In 2024, it might be "Low/Mid". The input number (50k) is the same, but the Label (High/Low) changed.
- **Fashion**: A style of shirt was "Cool" (Label=1) last year, now it is "Outdated" (Label=0).

## 4. Detection Methods

How do we know if drift is happening without labels?

**Statistical Tests** (Comparing Train vs Prod distributions):
1. **Kolmogorov-Smirnov (KS) Test**: Good for 1D continuous data. Checks if two samples come from same distribution.
2. **Chi-Squared Test**: For categorical data.
3. **Population Stability Index (PSI)**: Industry standard in finance.
4. **Kullback-Leibler (KL) Divergence**: Measures distance between distributions.

## 5. Hands-On: Simulating & Detecting Drift

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ks_2samp

# 1. Simulate Training Data (Normal Distribution)
np.random.seed(42)
train_data = np.random.normal(loc=50, scale=10, size=1000)

# 2. Simulate Prod Data (Drifted: Mean shifted by +5)
prod_data = np.random.normal(loc=55, scale=10, size=1000)

# 3. Visualize
plt.figure(figsize=(10, 5))
sns.kdeplot(train_data, label='Train (Reference)', fill=True)
sns.kdeplot(prod_data, label='Prod (Drifted)', fill=True)
plt.title("Data Drift Visualization: Mean Shift")
plt.legend()
plt.show()
```

```python
# 4. Detect with KS-Test
stat, p_value = ks_2samp(train_data, prod_data)

print(f"KS Statistic: {stat:.4f}")
print(f"P-Value: {p_value:.10f}")

if p_value < 0.05:
    print("ðŸš¨ DRIFT DETECTED! (Distributions are different)")
else:
    print("âœ… No Drift Detected.")

# Note: P-value < 0.05 means we reject the Null Hypothesis (that they are same)
```

## 6. Interview Preparation

**Q1: What is the difference between Data Drift and Concept Drift?**  
*A1: Data Drift is when input features change (e.g., users get younger). Concept Drift is when the relationship between input and target changes (e.g., inflation making salary less valuable).*

**Q2: How do you detect drift if you don't have labels yet?**  
*A2: You can only detect Data Drift (Covariate Shift) without labels by comparing input distributions using statistical tests like KS-Test or PSI. You cannot firmly confirm Concept Drift without delayed labels.*

**Q3: What do you do if you detect drift?**  
*A3: Retrain the model on the new data. If the drift is temporary (e.g., holiday spike), maybe use a separate holiday model.*
