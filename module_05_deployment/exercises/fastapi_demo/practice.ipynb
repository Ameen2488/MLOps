# Exercise: FastAPI Demo

**Objective**: Create a simple FastAPI app and run it.

*Note: Running FastAPI in a notebook requires `nest_asyncio` or running it in a background thread. For simplicity, we will write the code to a file and run it with `!uvicorn`.*

---

## 1. Write `app.py`

````python
%%writefile app.py
from fastapi import FastAPI
from pydantic import BaseModel

app = FastAPI()

class Numbers(BaseModel):
    a: int
    b: int

@app.post("/sum")
def sum_numbers(nums: Numbers):
    return {"result": nums.a + nums.b}

@app.get("/")
def home():
    return {"msg": "Hello from Notebook!"}
````

## 2. Run Server
Run this in a **Terminal**:
```bash
uvicorn app:app --reload
```
