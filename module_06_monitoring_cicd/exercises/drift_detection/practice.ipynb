# Exercise: Drift Detection

**Objective**: Detect drift in the Wine Quality dataset.

---

## 1. Load Data

```python
from sklearn.datasets import load_wine
import pandas as pd
from evidently.report import Report
from evidently.metric_preset import DataDriftPreset

data = load_wine(as_frame=True)
df = data.frame
```

## 2. Simulate Drift
Split the data into reference (first 100 rows) and current (last 78 rows).
The Wine dataset is sorted by class, so splitting by index WILL cause drift!

```python
# TODO: Create ref and curr dataframes
ref = df.iloc[:100]
curr = df.iloc[100:]
```

## 3. Generate Report
Use Evidently to detect the drift.

```python
# TODO: Run Report
report = Report(metrics=[DataDriftPreset()])
report.run(reference_data=ref, current_data=curr)
report.show_html()
```
