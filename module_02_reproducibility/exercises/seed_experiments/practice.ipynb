## 1. The "Unstable" Run
Run this cell multiple times. Does the accuracy change?

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

def train_unstable():
    X, y = load_iris(return_X_y=True)
    # No random_state!
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
    
    model = SVC()
    model.fit(X_train, y_train)
    return accuracy_score(y_test, model.predict(X_test))

print(f"Run 1: {train_unstable()}")
print(f"Run 2: {train_unstable()}")
```

## 2. The "Stable" Run
Fix the code below by adding `random_state=42` to `train_test_split`.

```python
def train_stable():
    X, y = load_iris(return_X_y=True)
    # TODO: Add random_state=42
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
    
    model = SVC()
    model.fit(X_train, y_train)
    return accuracy_score(y_test, model.predict(X_test))

print(f"Run 1: {train_stable()}")
print(f"Run 2: {train_stable()}")
```
