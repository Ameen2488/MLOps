## 1. Initialize

```python
import wandb
import random

# Ensure you are logged in via terminal: `wandb login`
wandb.init(project="exercise-wandb-practice")
```

## 2. Training Loop

```python
for epoch in range(10):
    loss = 1.0 / (epoch + 1) + random.random() * 0.1
    # TODO: Log loss
    wandb.log({"loss": loss, "epoch": epoch})
    
wandb.finish()
print("Check your W&B dashboard!")
```
